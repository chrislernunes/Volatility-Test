```python
import yfinance as yf
import matplotlib.pyplot as plt

# Function to calculate drawdown from a series of prices
def calculate_drawdown(prices):
    max_value = prices[0]
    drawdowns = []

    for price in prices:
        if price > max_value:
            max_value = price
        drawdown = (price - max_value) / max_value
        drawdowns.append(drawdown)

    return drawdowns

nifty_data = yf.download('^NSEI', start='2020-01-01')
drawdowns = calculate_drawdown(nifty_data['Close'])


plt.figure(figsize=(22, 8))
plt.plot(nifty_data.index, drawdowns, label='Nifty Drawdown', color='purple')
plt.axhline(0, color='black', linestyle='--', linewidth=1, label='Zero Line')

plt.title('Nifty Drawdown Chart')
plt.xlabel('Date')
plt.ylabel('Drawdown')
plt.legend()
plt.grid(True)
plt.show()
```

```python
niftybank_data = yf.download('^NSEBANK', start='2020-01-01')
drawdowns = calculate_drawdown(niftybank_data['Close'])


plt.figure(figsize=(22, 8))
plt.plot(niftybank_data.index, drawdowns, label='Nifty Bank Drawdown', color='purple')
plt.axhline(0, color='black', linestyle='--', linewidth=1, label='Zero Line')

plt.title('Nifty Bank Drawdown Chart')
plt.xlabel('Date')
plt.ylabel('Drawdown')
plt.legend()
plt.grid(True)
plt.show()
```
